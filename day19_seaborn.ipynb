# categorical
1. categorical plots(hue can be used) comparison with categorical data
   >countplot
   >barplot (requirement : one data categorical and one data numerical also calculates mean by default)
   >boxplot
   >violinplot
   >swarmplot
2. Distribution plot
   >distplot
   >displot
   >jointplot
   >pairplot
   >kdeplot
   >rugplot
3. Matrix plots
   >heatmap
   >clustermap
   >pairplot(same as above one)
4. Regression plot
   >lmplot
   >regplot
5. Grid plot
   >pairgrid
   >jointgrid
   >facetgrid

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set_style('whitegrid')
```

# displot

```python
sns.displot(data=tips,x='total_bill',y='tip')
```

```python
sns.displot(data=tips,x='total_bill')
```

# jointplot

```python
sns.jointplot(data=tips,x='total_bill',y='tip')
plt.show()

sns.jointplot(data=tips,x='total_bill',y='tip',hue='smoker')
plt.show()

sns.jointplot(data=tips,x='total_bill',y='tip',kind='hex')
plt.show()

sns.scatterplot(data=tips,x='total_bill',y='tip')
plt.show()
```

# kdeplot and rugplot

```python
fig=plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
plt.subplot(1,2,2)
sns.kdeplot(data=tips,x='total_bill',y='tip')
sns.rugplot(data=tips,x='total_bill',y='tip')
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

# Pairplot

```python
sns.pairplot(data=iris)
```

```python
sns.pairplot(data=iris,hue='species')
plt.show()
sns.pairplot(data=iris,hue='species',kind='reg')
plt.show()
```

# Regression plot

```python
sns.regplot(data=tips,x='total_bill',y='tip')
plt.show()
sns.lmplot(data=tips,x='total_bill',y='tip')
```

# Grid plot

```python
pg=sns.PairGrid(data=iris)
pg.map_diag(sns.kdeplot)
pg.map_lower(sns.scatterplot)
pg.map_upper(sns.kdeplot)
```

```python
jg=sns.JointGrid(data=tips,x='total_bill',y='tip')
jg.plot(sns.regplot,sns.distplot)
```

```python
fg=sns.FacetGrid(data=tips,row='smoker',col='day',hue='sex')
fg.map(sns.scatterplot,'total_bill','tip').add_legend()
```
