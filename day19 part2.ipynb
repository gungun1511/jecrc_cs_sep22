## Bivariate Analysis

### boxplot - cat,num

### jointplot - num,num
### scatterplot - num,num (x)

### barplot - num,num


### violinplot - cat,num
### swarmplot - cat,num(x)
### regplot - num,num(x)


### lineplot - num,num

### pointplot - num,num

### kdeplot-num,num
### rugplot-num,num(x)




```python
for col in cat_cols:
    for ncol in num_cols:
        plt.figure(figsize=(10,3))
        plt.suptitle(f'{col} vs {ncol}')
        plt.subplot(1,3,1)
        sns.boxplot(data=data,x=col,y=ncol,palette=choice(p_list))
        plt.subplot(1,3,2)
        sns.violinplot(data=data,x=col,y=ncol,palette=choice(p_list))
        plt.subplot(1,3,3)
        sns.barplot(data=data,x=col,y=ncol,palette=choice(p_list))
        plt.show()


```

```python
for col1 in num_cols:
    for col2 in num_cols:
        if col1 == col2:
            continue
        plt.figure(figsize=(10,3))
        plt.subplot(1,3,1)
        sns.jointplot(data=data,x=col1,y=col2 , kin='hex')
        plt.subplot(1,3,2)
        sns.pairplot(data=data,x=col1,y=col2 , kin='hex')
        plt.show()
```

```python
# sns.pairplot(data=data[num_cols])
pg=sns.PairGrid(data=data[num_cols])
pg.map_diag(plt.hist)
pg.map_lower(sns.regplot)
pg.map_upper(sns.kdeplot) 
```

```python
sns.heatmap(
    data=data[num_cols].corr(),
    annot=True,
    cmap=choice(p_list)

)
```

## Multivariate Analysis

```python
for col in cat_cols:
    if col =='Product':
        continue
    for ncol in num_cols:
        plt.figure(figsize=(10,3))
        plt.suptitle(f'{col} vs {ncol}')
        plt.subplot(1,3,1)
        sns.boxplot(data=data,x=col,y=ncol,hue='Product',palette=choice(p_list))
        plt.subplot(1,3,2)
        sns.violinplot(data=data,x=col,y=ncol,hue='Product',palette=choice(p_list))
        plt.subplot(1,3,3)
        sns.barplot(data=data,x=col,y=ncol,hue='Product',palette=choice(p_list))
        plt.show() 
```
